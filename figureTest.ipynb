```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 101)
print (x)
```

```python
x.shape
```

```python
x.size, type(x.size)
```

```python
def sinfunc(x):
    y = np.sin(x)
    return y
```

```python
y_sin = sinfunc(x)
y_sin
```

```python
plt.plot(x, y_sin)
```

```python
def cosfunc(x,  shifty = 0.0, mag = 1.0):
    y = np.cos(x) + shifty
    return y
```

```python
y_cos = cosfunc(x)
plt.plot(x, y_cos)
```

```python
plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label = "cos")
plt.legend()
plt.grid(True)
```

```python
fig, axes = plt.subplots(1, 2)
print(type(axes), axes.shape)
axes[0].plot(x, y_sin)
axes[1].plot(x, y_cos)
```

```python
y_cos0 = cosfunc(x, shifty=0.3, mag=1.0)
y_cos1 = cosfunc(x, shifty=-0.5, mag=1.5)
```

```python
plt.plot(x, y_sin, label = "sin")
plt.plot(x, y_cos, label = "cos")
plt.plot(x, y_cos0, label = "cos, shifty=0.3, mag=1.0")
plt.plot(x, y_cos1, label = "cos, shifty=-0.5, mag=1.5")
plt.legend()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
print(type(axes), "shape:", axes.shape, axes.size)
axes[0,0].plot(x, y_sin, label= "sin"); axes[0,0].legend()
axes[0,1].plot(x, y_cos, label= "cos"); axes[0,1].legend()
axes[1,0].plot(x, y_cos0, label = "cos0"); axes[1,0].legend()
axes[1,1].plot(x, y_cos1, label = "cos1"); axes[1,1].legend()
```

```python
axes.shape, axes.ravel().shape
```

```python
fig, axes = plt.subplots(2, 2)

for ax, y in zip(axes.ravel(), [y_sin, y_cos, y_cos0, y_cos1]):
    print(type(ax), type(y), y.shape)
    ax.plot(x, y)
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```
