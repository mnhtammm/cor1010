# 1. Logistic curve

```python
def f(x, L, k, x0):
    y = L / (1 + np.exp( -k * (x-x0)))
    return y
```

```python
L = 1. #ghi them dau (.) dang sau de define no laf float chu khong phai la int
k = 1.
x0 = 0.
xvalues = np.linspace(-6, 6, 500)
yvalues = []
for x in xvalues:
    y = f(x, L, k, x0)
    yvalues.append(y)
yvalues
```

```python
plt.plot(xvalues,yvalues)
plt.title("Logistic curve 1.1")
plt.xlabel("xlable")
plt.ylabel("ylable")
```

```python
L = 1
k = 1
x0 = 2
xvalues = np.linspace(-6, 6, 500)
yvalues = []
for x in xvalues:
    y = f(x, L, k, x0)
    yvalues.append(y)
plt.plot(xvalues,yvalues, "--")
plt.title("Logistic curve 1.2")
plt.xlabel("xlable")
plt.ylabel("ylable")
```
