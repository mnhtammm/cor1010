```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lineEquation(x, a, b):
    y = a * x + b
    return y
```

```python
y = lineEquation(x=1, a=2, b=3)
print(y)
```

```python
a = 2
b = 3

xvalues = [-1, 0, 1, 2 , 3]
yvalues = []
for x in xvalues:
    y = lineEquation(x,a,b)
    yvalues.append(y)
```

```python
yvalues
```

```python
plt.plot(xvalues, yvalues, 'o-')
```

```python
# 3.2 **2 = (3.2)^2 = np.power(3.2, 2)
```

```python
def bac2(x, a, b, c):
    y = a * np.power(x, 2) + b * x + c
    return y
```

```python
a = 1
b = 2
c = 3

xvalues = np.linspace(-2, 4, 100) #type(xvalue) = numpy.ndarray
yvalues = []
for x in xvalues:
    y = bac2(x, a, b, c)
    yvalues.append(y)
plt.plot(xvalues, yvalues, '--',color = "pink")
```
